```python
 # K-Means clustering
# Fit K-Means clustering algorithm with 3 clusters
#Importing the necessary libraries
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Extract features to be used for clustering
data = pd.read_csv("C:\\Users\\n\\Downloads\\Country-data.csv")
X = data.drop([ 'country', 'gdpp'], axis=1)
# Standardize the data
scaler = StandardScaler()
X_std = scaler.fit_transform(X)
```

```python

kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X_std)

# Assign clusters to each data point
data['Cluster'] = kmeans.labels_

# Print cluster centers
print(kmeans.cluster_centers_)
```

```python
#Expectation Maximization:
# A strategy for carrying out maximum likelihood estimation in
# the presence of latent variables is the expectation-maximization algorithm.
# To achieve this, the model is first optimized, then the latent variable values are estimated, 
# and so on until convergence.
```

```python
#Importing the necessary libraries
import pandas as pd
import numpy as np
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler
```

```python
# Extract features to be used for clustering
X = data.drop([ 'country', 'gdpp'], axis=1)

# Standardize the data
scaler = StandardScaler()
X_std = scaler.fit_transform(X)
```

```python
# Import libraries
import pandas as pd
import numpy as np
# import matplotlib.pyplot as plt
# from sklearn.mixture import GaussianMixture
# from sklearn.cluster import KMeans, GaussianMixture
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
# Fit Expectation Maximization algorithm with 3 clusters
em = GaussianMixture(n_components=3, covariance_type='full', random_state=42)
em.fit(X_std)

# Assign clusters to each data point
data['Cluster'] = em.predict(X_std)

# Print cluster means
print(em.means_)
```

```python
#PCA for Country Dataset
#importing the necessary libraries
from sklearn.decomposition import PCA
import pandas as pd
```

```python
# Load data
data = pd.read_csv("C:\\Users\\n\\Downloads\\Country-data.csv")
# Separate features and labels
X = data.drop(['country'], axis=1)
y = data['country']

# Apply PCA
pca = PCA(n_components=5)
X_pca = pca.fit_transform(X)

# Save PCA results to new CSV file
result = pd.DataFrame(X_pca, columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5'])
result['country'] = y
result.to_csv('Country_data_PCA.csv', index=False)
print(result)
```
